### 1. Load Matlab Simulated data DataSinFreq

```python
import scipy.io as scio

data = scio.loadmat(r'DataSinFreq.mat')
print(data)
```

#### 1 1.1 Visualisation des données

```python
X = data['X']
y = data['target']
print("Nombre d'observations: ",X.shape[0])
print("Nombre de features: ",X.shape[1])
print("Shape target: ",y.shape )
```

```python
#Check if there are nan values
import numpy as np
arr = np.sum(X)
is_nan = np.isnan(arr)
print(is_nan)
#No nan values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,5))
plt.plot(X[:,:]);
plt.title("Visualisation des observations");
plt.xlabel("Feature")
plt.ylabel("Value")
plt.show()
```

### 2 . Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
preprocessor = make_pipeline(StandardScaler(),PCA(n_components = 0.99,random_state=0))
```

```python
X_train_pca = preprocessor.fit_transform(X_train,y_train)
X_test_pca = preprocessor.transform(X_test)
X_train_pca.shape
```

```python
plt.figure(figsize=(5,5))
CompInit = 0
CompFinal = 1
plt.scatter(X_train_pca[:,CompInit],X_train_pca[:,CompFinal],c = y_train);
plt.scatter(X_test_pca[:,CompInit],X_test_pca[:,CompFinal],c='r')
plt.colorbar();
plt.title("Visualisation des données d'entrainement et de test (en rouge)")
plt.xlabel("Componante n{0}".format(CompInit))
plt.ylabel("Componante n{0}".format(CompFinal))


plt.figure(figsize=(5,5))
CompInit = 0
CompFinal = 1
plt.scatter(X_test_pca[:,CompInit],X_test_pca[:,CompFinal],c =y_test);
plt.colorbar();
plt.title("Visualisation des données de test")
plt.xlabel("Componante n{0}".format(CompInit+1))
plt.ylabel("Componante n{0}".format(CompFinal+1))
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import VotingRegressor 
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.exceptions import DataConversionWarning
```

```python
seed = 0
models = []
models.append(('LR',make_pipeline(StandardScaler(),LinearRegression())))
models.append(('Ridge',make_pipeline(StandardScaler(),Ridge(random_state=0))))
#models.append(('Lasso',make_pipeline(StandardScaler(),Lasso(random_state=0))))
#models.append(('ElasticNet',make_pipeline(StandardScaler(),ElasticNet(random_state=0))))
models.append(("SVR",make_pipeline(StandardScaler(),SVR())))
models.append(("CART",make_pipeline(StandardScaler(),DecisionTreeRegressor(random_state=0))))
models.append(("K-Neighbor",make_pipeline(StandardScaler(),KNeighborsRegressor())))
```

```python
pipelines_ensemble = []
pipelines_ensemble.append(("Voting",make_pipeline(StandardScaler(),VotingRegressor(models))))
pipelines_ensemble.append(("Bagging",make_pipeline(StandardScaler(),BaggingRegressor(DecisionTreeRegressor()))))
pipelines_ensemble.append(("RandomForest",make_pipeline(StandardScaler(),RandomForestRegressor())))
pipelines_ensemble.append(("ExtraTrees",make_pipeline(StandardScaler(),ExtraTreesRegressor())))
#pipelines_ensemble.append(("Adaboost",make_pipeline(StandardScaler(),AdaBoostRegressor())))
pipelines_ensemble.append(("GradientBoost",make_pipeline(StandardScaler(),GradientBoostingRegressor())))
pipelines_ensemble.append(("Stacking",make_pipeline(StandardScaler(),StackingRegressor(models,final_estimator=KNeighborsRegressor()))))
```

```python
pipelines_other = []
pipelines_other.append(("XGB",make_pipeline(StandardScaler(),XGBRegressor())))
pipelines_other.append(("LBGM",make_pipeline(StandardScaler(),LGBMRegressor())))
```

```python
pipelines = models + pipelines_ensemble + pipelines_other
```

```python
import warnings
warnings.filterwarnings("ignore", category=DataConversionWarning)
```

```python
results=[]
names=[]
num_folds=5
scoring='r2'

for name, model in pipelines:
    kfold=KFold(n_splits=num_folds)
    cv_results=cross_val_score(model, X_train_pca,y_train,cv=kfold,scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name,cv_results.mean(),cv_results.std())
    print(msg)
```

```python
plt.figure(figsize=(20,5))
plt.boxplot(results);
tick = [i for i in range(1,len(names)+1)]
plt.xticks(tick,names);
plt.title('Algorithm Comparison');
```

```python
a = dict(zip(names,results))
for i in a:
    a[i] = a[i].mean()
print(a)
```
